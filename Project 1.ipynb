```python
import pandas as pd
import os
import numpy as np
from config import geoapify_key
from aus_config import auspost_key
import requests
import json

data_1="2016_population_data.csv"
data_2="2036_population_data.csv"

current_data=pd.read_csv(data_1)
future_data=pd.read_csv(data_2)

growth_data=pd.merge(current_data,future_data,how="left",on="Area Name")
growth_data.head()
```

```python
new_df=pd.DataFrame(growth_data)
new_df1=new_df.drop("GCCSA_y", axis='columns')
new_df1.head()
```

```python
new_df2=new_df1.rename(columns={ "GCCSA_x":"Area Type", "Total_x":"Total_2016","Total_y":"Total_2036"})
new_df2.head()
```

```python
total_areas=growth_data["Area Name"].nunique()
total_areas
```

```python
new_df2["Total_2016"]=new_df2["Total_2016"].apply(np.int64)
print(new_df2.dtypes)
```

```python
new_df2["Population Difference"]=new_df2['Total_2036'] - new_df2['Total_2016']
new_df2["Population Growth"]=(new_df2['Total_2036'] - new_df2['Total_2016'])/new_df2['Total_2016']*100
new_df2
```

```python
suburbarray = new_df2["Area Name"].unique()
listofsuburbs = suburbarray.tolist()

# firstSuburb = "Carlton"
# target_url = f"https://auspostie.com/{firstSuburb}.json"
# results = requests.get(target_url).json()


for suburb in listofsuburbs:
    found=False
    target_url = f"https://auspostie.com/{suburb}.json"
    try:
        results = requests.get(target_url).json()
        for x in results:
            if x['state'] == "VIC" and found==False:
                found=True
            else:
                next
        print(f"was a postcode was found for {suburb}? -{found}")
    except ValueError:
        print(f"The suburb {suburb} was unable to be searched")

# place_id = results["results"][0]["postcode"]
```
